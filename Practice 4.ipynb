````python
#156 Binary Tree Upside Down
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    
    def upsideDownBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        if not root:
            return root
        stack=[]
        while root:
            stack.append(root)
            root=root.left
        new_root=stack.pop()
        while len(stack)>0:
            parent=stack.pop()
            parent.left=None
            new_root.left=parent.right
            new_root.right=parent
            parent.right=None
            new_root=new_root.right
        return new_root
class TreeNode(object):
    def __init__(self,x):
        self.val=x
        self.left=None
        self.right=None
        
    
    
````
